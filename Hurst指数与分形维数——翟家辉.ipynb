```python
#å¸¸ç”¨åº“åŠå‘½ä»¤

import numpy as np
    # çŸ©é˜µè¿ç®—åº“
import matplotlib.pyplot as plt
    # ç»˜å›¾
import ipywidgets as widgets
    # è¾“å…¥æ§ä»¶
import matplotlib.gridspec as gridspec
    # è‡ªå®šä¹‰å›¾å½¢å¸ƒå±€
import random
    # ç”Ÿæˆéšæœºæ•°
import math
    # æ•°å­¦è¿ç®—åŒ…
import scipy as sc
    # å½¢å¼è¿ç®—æ±‚è§£æè§£
from scipy.optimize import curve_fit
    # å‡½æ•°æ‹Ÿåˆ
from scipy.fftpack import fft
    # å¿«é€ŸFFTå˜æ¢
%matplotlib notebook
    # ç»˜å›¾æ¨¡å¼
plt.rcParams['font.sans-serif'] = ['SimHei']
    # ç»˜å›¾æ˜¾ç¤ºä¸­æ–‡
plt.rcParams['axes.unicode_minus'] = False
    # ç»˜å›¾æ˜¾ç¤ºè´Ÿå·
```

```python
# Zplt è‡ªå®šä¹‰ç»˜å›¾å‡½æ•°ï¼Œå‚æ•°è§é™„å½•
def Zplt(X,Y,Z=0,Box=0,Size=[9,9],Dim=2,Bm=0,Type="P",La="-",Lx="X",Ly="Y",Lz="Z",Tn="",Ls=15,Lw=1,Lm="-",Lc="black",Al=1,Ms=2,Mm="",Mc="black",AE=[30,-60],Re=[],**Other):

    if Bm == 0:# æ­£å¸¸æƒ…å†µï¼Œå•å›¾ç»˜åˆ¶
        if Size != True:
            plt.figure(figsize=Size)
            plt.title(Tn,size=Ls)
            plt.xlabel(Lx,size=Ls)
            plt.ylabel(Ly,size=Ls)
        if Dim == 2:# äºŒç»´å›¾
            
            
            if Type == "P":# plot
                
                if Mm == "":# æ— æ ‡å¿—ç‚¹
                    plt.plot(X,Y,color=Lc,linewidth=Lw,alpha=Al,label=La)
                
                else:# æœ‰æ ‡å¿—ç‚¹
                    plt.plot(X,Y,color=Lc,linewidth=Lw,alpha=Al,marker=Mm,markersize=Ms,markerfacecolor=Mc,label=La)
                
            elif Type == "S": # scatter
                plt.scatter(X,Y,s=Ms,c=Mc,marker=Mm,alpha=Al,label=La)
                
            elif Type == "H":# hist
                n,bins,patches=plt.hist(X,Y,ec=Lc,fc=Mc,density=Lm,alpha=Al,label=La)
                
            plt.legend(loc="best")
            
        elif Dim == 3: 
            plt.view_init(elev=AE[0],azim=AE[1])
            
            if Size != True:
                plt.axes(projection='3d')
                plt.xlabel(Lx,size=Ls)
                plt.ylabel(Ly,size=Ls)
                plt.zlabel(Lz,size=Ls)
                
            if Type == "P":# plot
                
                if Mm == "":# æ— æ ‡å¿—ç‚¹
                    plt.plot(X,Y,Z,color=Lc,linewidth=Lw,alpha=Al,label=La)
                
                else:# æœ‰æ ‡å¿—ç‚¹
                    plt.plot(X,Y,Z,color=Lc,linewidth=Lw,alpha=Al,marker=Mm,markersize=Ms,markerfacecolor=Mc,label=La)
                
            elif Type == "S": # scatter
                plt.scatter(X,Y,s=Ms,c=Mc,marker=Mm,alpha=Al,label=La)
                
            elif Type == "H":# hist
                n,bins,patches=plt.hist(X,Y,ec=Lc,fc=Mc,density=Lm,alpha=Al,label=La)
                
            plt.legend(loc="best")            
        
    elif Bm == 1:# åˆ†æ ¼æ§åˆ¶
        plt.figure(figsize=Size)
        global gs
        gs = gridspec.GridSpec(Box[0],Box[1])
        plt.suptitle(Tn,size=Ls)       
        
    else:# å›¾ä¸­å›¾
        Bm[4]=eval(f"plt.subplot(gs[{Bm[0]}:{Bm[1]},{Bm[2]}:{Bm[3]}])")
        
        if Dim == 2:# äºŒç»´å›¾
            Bm[4].xlabel(Lx,size=Ls)
            Bm[4].ylabel(Ly,size=Ls)
            
            if Type == "P":# plot
                
                if Mm == "":# æ— æ ‡å¿—ç‚¹
                    Bm[4].plot(X,Y,color=Lc,linewidth=Lw,alpha=Al,label=La)
                
                else:# æœ‰æ ‡å¿—ç‚¹
                    Bm[4].plot(X,Y,color=Lc,linewidth=Lw,alpha=Al,marker=Mm,markersize=Ms,markerfacecolor=Mc,label=La)
                
            elif Type == "S": # scatter
                Bm[4].scatter(X,Y,s=Ms,c=Mc,marker=Mm,alpha=Al,label=La)
                
            elif Type == "H":# hist
                n,bins,patches=Bm[4].hist(X,Y,ec=Lc,fc=Mc,density=Lm,alpha=Al,label=La)
                
            Bm[4].legend(loc="best")
            
        elif Dim == 3: 
            
            if Size != True:
                Bm[4].axes(projection='3d')
                Bm[4].xlabel(Lx,size=Ls)
                Bm[4].ylabel(Ly,size=Ls)
                Bm[4].zlabel(Lz,size=Ls)
                
            if Type == "P":# plot
                
                if Mm == "":# æ— æ ‡å¿—ç‚¹
                    Bm[4].plot(X,Y,Z,color=Lc,linewidth=Lw,alpha=Al,label=La)
                
                else:# æœ‰æ ‡å¿—ç‚¹
                    Bm[4].plot(X,Y,Z,color=Lc,linewidth=Lw,alpha=Al,marker=Mm,markersize=Ms,markerfacecolor=Mc,label=La)
                
            elif Type == "S": # scatter
                Bm[4].scatter(X,Y,s=Ms,c=Mc,marker=Mm,alpha=Al,label=La)
                
            elif Type == "H":# hist
                n,bins,patches=Bm[4].hist(X,Y,ec=Lc,fc=Mc,density=Lm,alpha=Al,label=La)
                
            Bm[4].legend(loc="best")
            
    if len(Other) != 0:# é¢å¤–å‘½ä»¤
        for i in range(1,len(Other)+1):
            exec(Other[f'o{i}'])
            
    #è¿”å›å€¼
    if Re == 'H':
        return [n,bins,patches]# é¢‘ç‡æˆ–é¢‘æ•°åˆ†å¸ƒ
    
    elif Re == 0:
        Zplt="HUSKY"
        print("QAQ,ä»£ç ç ä¸åŠ¨äº†")
        
    elif Re != []:
        R=eval(Re)
        return R
        
```

<font size=3.5>è¿™äº›æ–¹æ³•ä¸­ï¼Œæ¯ç§æ–¹æ³•çš„é€‚ç”¨èŒƒå›´å’Œä¼˜åŠ£åŠ¿å„ä¸ç›¸åŒï¼Œåœ¨æœ¬æ–‡ä¸­æˆ‘ä»¬é‡‡ç”¨é«˜æ–¯è¿‡ç¨‹ç”Ÿæˆæ³•ä¸‹çš„**éšæœº Weierstrass å‡½æ•°**ç”Ÿæˆ fBM ã€‚

>### d. éšæœºé­å°”æ–¯ç‰¹æ‹‰æ–¯å‡½æ•°ç”Ÿæˆ fBM

<font size=3.5>**Weierstrasså‡½æ•°**æ˜¯ä¸€ç±»å¤„å¤„è¿ç»­è€Œå¤„å¤„ä¸å¯å¯¼çš„å®å€¼å‡½æ•°ã€‚ç»™å®šå‚æ•° $b>1$ å’Œ $0<H<1$ï¼Œå®šä¹‰å‡½æ•° $W(t)$ ä¸ºï¼š

<font size=3.5>$$W\bigl(t\bigr)=\sum_{k=1}^{\infty}c_{k}b^{-H k}\sin\bigl(b^{k}t+d_{k}\bigr)$$

<font size=3.5>å…¶ä¸­ $c_k$ å±äºå‡å€¼ä¸º 0 ã€æ ‡å‡†å·®ä¸º 1 çš„æ­£æ€åˆ†å¸ƒï¼Œ$d_k$ åœ¨ $0â‰¤d_k<2Ï€$ çš„åŒºé—´å†…å‡åŒ€åˆ†å¸ƒã€‚åˆ©ç”¨ Weierstrass å‡½æ•°å³å¯ç”Ÿæˆå…·æœ‰ç‰¹å®š $Hurst$ æŒ‡æ•°çš„ fBm æ•°æ®ã€‚

```python
def weierstrass_function(x,Hurst_exponent,b):
    """
    è®¡ç®—éšæœº Weierstrass å‡½æ•°ï¼Œç”ŸæˆfBMæ•°æ®
    
    å‚æ•°ï¼š
    x - æ—¶é—´åºåˆ—
    Hurst_exponent - Hurst æŒ‡æ•°
    b - Weierstrass å‡½æ•°çš„ç±»å‘¨æœŸæ§åˆ¶å‚é‡
    
    è¿”å›ï¼š
    å¯¹åº”xçš„Weierstrass å‡½æ•°å€¼
    
    """
    s = 0
    random.seed()
    #å–å‰100é¡¹
    c=np.random.standard_normal(1000)
    d=np.random.uniform(0,2*np.pi,size=1000)
    for k in range(1000):
        s += c[k] * np.power(b,-(k+1) * Hurst_exponent) * np.sin(  np.power(b,k+1) * x + d[k])
    return s
```

## 2. *èµ«æ–¯ç‰¹æŒ‡æ•°ï¼ˆHurst Exponentï¼‰*

<font size=3.5>**Hurst æŒ‡æ•°** $H$ è¢«ç”¨äºåˆ»ç”»æ•°æ®é•¿è®°å¿†æ€§ï¼›å®ƒå¸¸è¢«ç”¨æ¥æµ‹é‡ä¸€ä¸ªæ—¶é—´åºåˆ—çš„æ³¢åŠ¨èŒƒå›´å¦‚ä½•éšæ—¶é—´è·¨åº¦å˜åŒ–è€Œå˜åŒ–ã€‚1951å¹´ï¼Œè‹±å›½æ°´æ–‡å­¦å®¶èµ«æ–¯ç‰¹ï¼ˆHarold Edwin Hurstï¼‰åœ¨ç ”ç©¶å°¼ç½—æ²³æ°´ä½å˜åŒ–æ—¶å‘ç°äº†æ—¶é—´åºåˆ—ä¸­å­˜åœ¨çš„**é•¿è®°å¿†æ€§**ï¼Œå³æ—¶é—´åºåˆ—å½“å‰æˆ–è¿‡å»çš„å–å€¼ä»¥è¿œè¶…éšæœºæ‰°åŠ¨æ‰€èƒ½è¾¾åˆ°çš„ç¨‹åº¦å½±å“è¯¥æ—¶é—´åºåˆ—åœ¨æœªæ¥çš„å–å€¼ã€‚å¦‚æœä¸€ä¸ªå¹³ç¨³æ—¶é—´åºåˆ—çš„è‡ªç›¸å…³å‡½æ•° $p(k)$ çš„è¡°å‡é€Ÿåº¦æœä»**å¹‚å¾‹è¡°å‡**ï¼ˆå³æ¯”æŒ‡æ•°è¡°å‡æ…¢ï¼‰ï¼Œé‚£ä¹ˆè¿™ä¸ªæ—¶é—´åºåˆ—å°±å…·å¤‡é•¿è®°å¿†æ€§ã€‚è®°å¿†æ€§ä½“ç°åœ¨è‡ªç›¸å…³å‡½æ•°çš„éç‹¬ç«‹æ€§ä¸Šï¼Œè€Œâ€œé•¿â€ä½“ç°åœ¨è¡°å‡çš„é€Ÿåº¦æ…¢ã€‚

>### a. Hurst æŒ‡æ•°çš„æ¥æº

<font size=3.5>å“ˆç½—å¾·Â·åŸƒå¾·æ¸©Â·èµ«æ–¯ç‰¹æ˜¯ä¸€ä½æ°å‡ºçš„è‹±å›½æ°´æ–‡å­¦å®¶, ä»–æ›¾å‚ä¸å°¼ç½—æ²³å¤§åå·¥ç¨‹ã€‚åœ¨æ–½å·¥è¿‡ç¨‹ä¸­, èµ«æ–¯ç‰¹éœ€è¦è¯„ä¼°æ°´ä½çš„æ³¢åŠ¨ã€‚æœ€åˆ, ä»–å‡è®¾æ°´æµæ˜¯éšæœºæ³¨å…¥çš„éšæœºè¿‡ç¨‹ã€‚ç„¶è€Œ, èµ«æ–¯ç‰¹åœ¨ç ”ç©¶äº†å°¼ç½—æ²³ä¹ä¸ªä¸–çºªçš„æ´ªæ°´è®°å½•åæ€»ç»“å‡ºï¼šé«˜äºå¹³å‡æ°´å¹³çš„æ´ªç¾éšåä¼šæœ‰æ›´å¼ºåŠ²çš„æ´ªç¾ã€‚ä¹‹åè¿™ä¸ªè¿‡ç¨‹æ”¹å˜å…¶æ–¹å‘, ä½äºå¹³å‡æ°´å¹³çš„æ´ªç¾ä¼šç´§éšæ›´å¼±çš„æ´ªç¾ã€‚æ˜¾ç„¶è¿™äº›æ˜¯å±äºéå‘¨æœŸæ€§çš„è½®å›ã€‚

<font size=3.5>èµ«æ–¯ç‰¹çš„ç»Ÿè®¡æ¨¡å‹åŸºäºé˜¿å°”ä¼¯ç‰¹Â·çˆ±å› æ–¯å¦å…³äºå¸ƒæœ—è¿åŠ¨çš„ç²’å­çš„éšæœºæ•£å¸ƒæ¨¡å‹ã€‚è¿™ä¸ªç†è®ºèƒŒåçš„æ€è·¯æ˜¯, ç²’å­èµ°è¿‡çš„è·ç¦» $(R)$ ä¸æ—¶é—´ $(T)$ çš„å¹³æ–¹æ ¹æˆæ¯”ä¾‹å¢åŠ ,å³ $R=T^{0.5}$ã€‚åœ¨æ­¤ç ”ç©¶è¿‡ç¨‹ä¸­èµ«æ–¯ç‰¹æ‰©å±•äº†çˆ±å› æ–¯å¦çš„æ–¹ç¨‹, å°†å…¶è½¬åŒ–ä¸ºæ›´å¸¸ç”¨çš„å½¢å¼:

<font size=3.5>$$(R/S)_n=C\times n^H$$

<font size=3.5>æ­¤å¤„ $c$ æ˜¯ä¸€ä¸ªå¸¸æ•°ã€‚

>### b. Hurst æŒ‡æ•°åº¦é‡ç²—ç³™åº¦

```python
def weierstrass_function_H(x,Hurst_exponent,b,c,d):
    """
    è®¡ç®—éšæœº Weierstrass å‡½æ•°ï¼Œç”ŸæˆfBMæ•°æ®
    
    å‚æ•°ï¼š
    x - æ—¶é—´åºåˆ—
    Hurst_exponent - Hurst æŒ‡æ•°
    b - Weierstrass å‡½æ•°çš„ç±»å‘¨æœŸæ§åˆ¶å‚é‡
    c,d - Weierstrass å‡½æ•°çš„éšæœºå‚é‡
    
    è¿”å›ï¼š
    å¯¹åº”xçš„Weierstrass å‡½æ•°å€¼
    
    """
    s = 0
    for k in range(len(c)):
        s += c[k] * np.power(b,-(k+1) * Hurst_exponent) * np.sin( np.power(b,k+1) * x + d[k])
    return s
```

```python
#
def update_H(frame):
    """
    æ›´æ–°å¸§
    
    å‚æ•°ï¼š
    frame - å¸§
    
    è¿”å›ï¼š
    å¯¹åº”frameçš„å½“å‰å¸§ç»˜å›¾
    
    """
    global c,d,b
    x = np.linspace(0,10,1000)
    y = weierstrass_function_H(x,frame,b,c,d)
    ax.set_xlim(0,10)
    ax.set_ylim(-7,7)
    ax.set_xlabel('t',size=16)
    ax.set_ylabel('W(t)',size=16)
    ln.set_data(x,y)
    text_pt.set_text('ä¸åŒHä¸‹çš„æ•°æ®ç²—ç³™åº¦ ( H = %.2f ,æ‰§è¡ŒåŠ¨ç”»è¯·é‡è¿è¡Œæœ¬Cell)' %(frame))
    return ln,


random.seed()
c=np.random.standard_normal(100)
d=np.random.uniform(0,2*np.pi,size=100)
b=np.random.uniform(1,5)

data_gen=np.linspace(0,1,100)
fig=plt.figure(figsize=(10,4),dpi=80)
ax=fig.add_subplot(111)
ln, = plt.plot(np.array([]), np.array([]),color="c",linewidth=2,animated=False)
text_pt=plt.title('',fontsize=12)
anim = animation.FuncAnimation(fig, update_H, frames=data_gen,interval=200,init_func=init,blit=True,repeat=False)
anim.save('weierstrass_function_show_2.gif',writer='pillow')#ä¿å­˜ä¸ºgifå½¢å¼
```

<font size=3.5>**åˆ†å½¢ç›´æ–¹å›¾æ³•**ï¼šä½¿ç”¨ä¸åŒçš„å°ºåº¦æ¥æ„å»ºåˆ†å½¢ç›´æ–¹å›¾ï¼Œä»¥è¯„ä¼°æ•°æ®çš„åˆ†å½¢ç»´æ•°ã€‚è¯¥æ–¹æ³•ä¸å‡ ä½•ç›´æ–¹å›¾æ³•ç±»ä¼¼ï¼Œä½†å®ƒè€ƒè™‘äº†åˆ†å½¢ç‰¹å¾çš„ç©ºé—´å…³ç³»ã€‚

<font size=3.5>**å¤šåˆ†å½¢ç»´æ•°æ–¹æ³•**ï¼ˆMFDï¼‰ï¼šä½¿ç”¨ä¸€ç»„ä¸åŒçš„åˆ†è£‚æ¯”ç‡å¯¹æ•°æ®è¿›è¡Œåˆ†æï¼Œä½¿ç”¨å¯¹æ•°å¯¹æ•°å›¾è¯„ä¼°ç»“æœã€‚åœ¨ MFD æ–¹æ³•ä¸­ï¼Œæ¯ä¸ªåˆ†è£‚æ¯”ç‡éƒ½è¢«ç”¨æ¥å¯¹æ•°æ®è¿›è¡Œåˆ†æï¼Œå¹¶ä½¿ç”¨å¯¹æ•°å¯¹æ•°å›¾è¯„ä¼°ç»“æœã€‚åˆ†å½¢ç»´æ•°æ˜¯æ‰€æœ‰åˆ†è£‚æ¯”ç‡çš„è¶‹åŠ¿çº¿çš„æ–œç‡çš„å¹³å‡å€¼ã€‚

>### b. è®¡ç›’ç»´æ•°

<font size=3.5>åœ¨åˆ†å½¢å‡ ä½•ä¸­, **[è®¡ç›’ç»´æ•°](https://baike.baidu.com/item/%E8%AE%A1%E7%9B%92%E7%BB%B4%E6%95%B0/22781986)**ä¹Ÿç§°ä¸º **ç›’ç»´æ•°** ã€*é—µå¯å¤«æ–¯åŸºç»´æ•°*ï¼Œæ˜¯ä¸€ç§æµ‹é‡è·ç¦»ç©ºé—´(X,d)ï¼ˆç‰¹åˆ«æ˜¯è±ªæ–¯å¤šå¤«ç©ºé—´ï¼‰æ¯”å¦‚æ¬§æ°ç©ºé—´Rä¸­åˆ†å½¢ç»´æ•°çš„è®¡ç®—æ–¹æ³•ã€‚è¦è®¡ç®—åˆ†å½¢Sçš„ç»´æ•°ï¼Œä½ å¯ä»¥æƒ³è±¡ä¸€ä¸‹æŠŠè¿™ä¸ªåˆ†å½¢æ”¾åœ¨ä¸€ä¸ªå‡åŒ€åˆ†å‰²çš„ç½‘æ ¼ä¸Šï¼Œæ•°ä¸€æ•°æœ€å°éœ€è¦å‡ ä¸ªæ ¼å­æ¥è¦†ç›–è¿™ä¸ªåˆ†å½¢ã€‚é€šè¿‡å¯¹ç½‘æ ¼çš„é€æ­¥ç²¾åŒ–ï¼ŒæŸ¥çœ‹æ‰€éœ€è¦†ç›–æ•°ç›®çš„å˜åŒ–ï¼Œä»è€Œè®¡ç®—å‡ºè®¡ç›’ç»´æ•°ã€‚
    
<font size=3.5>å‡è®¾å½“æ ¼å­çš„è¾¹é•¿æ˜¯$r$æ—¶ï¼Œæ€»å…±æŠŠç©ºé—´åˆ†æˆNä¸ªæ ¼å­ï¼Œé‚£ä¹ˆè®¡ç›’ç»´æ•°å°±æ˜¯ï¼š$$ D = \lim_{n \to 0} \frac{\text{log}(N_{r})}{\text{log}(1/r)} $$

<font size=3.5>å½“æé™ä¸æ”¶æ•›æ—¶ï¼Œæˆ‘ä»¬å¿…é¡»æŒ‡å‡º **é¡¶ç›’ç»´æ•°** æˆ– **åº•ç›’ç»´æ•°** ï¼Œæˆ–è€…è¯´ï¼Œè®¡ç›’ç»´æ•°ä»…åœ¨å’Œé¡¶ç›’ç»´æ•°ä¸åº•ç›’ç»´æ•°ç›¸ç­‰æ—¶æ‰æ˜¯æœ‰å®šä¹‰çš„ã€‚é¡¶ç›’ç»´æ•°ä¹Ÿç§°ä¸ºèƒ½é‡ç»´æ•°ã€ç§‘è«æ ¼æ´›å¤«ç»´æ•°ã€ç§‘è«æ ¼æ´›å¤«å®¹ç§¯ï¼Œæˆ–è€…é—µå¯å¤«æ–¯åŸºä¸Šç•Œç»´æ•°ï¼Œç±»ä¼¼çš„å¯å®šä¹‰é—µå¯å¤«æ–¯åŸºä¸‹ç•Œç»´æ•°ã€‚</font>

>### c. è®¡ç›’ç»´æ•°çš„åˆ†æ ¼å±•ç¤º

```python

def BOX(box_N): # MAX 100,000 å¾—äºŒç»´çŸ©é˜µ
    """
    ç”Ÿæˆç”¨äºè®¡ç®—è®¡ç›’ç»´æ•°äºŒç»´çŸ©é˜µ
    
    å‚æ•°ï¼š
    box_N - æ–¹é˜µè¾¹é•¿
    
    è¿”å›ï¼š
    box_B - box_N*box_Nçš„æ–¹é˜µ
    
    """
    box_B=np.zeros((box_N+1,box_N+1),dtype=np.bool_)
    return box_B
```

```python
#
def Determine_W(step_N,Hurst):  
    """
    ç”ŸæˆWå‡½æ•°è½¨è¿¹ç‚¹æ•°æ®ä»¥åŠå¸ƒå°”åˆ¤æ®åˆ—
    
    å‚æ•°ï¼š
    step_N - è®¡ç®—æ­¥æ•°
    Hurst - å¯¹åº”Hurstå€¼
    
    è¿”å›ï¼š
    XY0 - Xï¼ŒYï¼Œå¸ƒå°”ä¸‰è¡ŒçŸ©é˜µ
    
    """
    x0=np.linspace(0,1,step_N)
    y0=np.zeros((2,step_N))
    y0[0]=weierstrass_function(x0,Hurst,1.2)
    XY0=np.vstack([x0, y0])

    return XY0

def Determine_W_show(step_N,box_N,Hurst):# ç‚¹é˜µå›¾
    """
    ç”ŸæˆWå‡½æ•°è½¨è¿¹ç‚¹çš„åˆ†æ ¼å±•ç¤º
    
    å‚æ•°ï¼š
    step_N - è®¡ç®—æ­¥æ•°
    Hurst - å¯¹åº”Hurstå€¼
    box_N - æ–¹é˜µè¾¹é•¿
    
    è¿”å›ï¼š
    ç»˜å›¾
    
    """    
    Plot_black=Determine_W(step_N,Hurst)
    Plot_black[1]=Plot_black[1]/(np.max(Plot_black[1])-np.min(Plot_black[1]))
    Zplt(Plot_black[0],Plot_black[1],La="åˆ†æ•°å¸ƒæœ—è¿åŠ¨",Al=0.6)
    plt.title("è®¡ç›’ç»´æ•°çš„åˆ†æ ¼å±•ç¤º")
    BOX_k=BOX(box_N)
    
# ç¡®å®šæ ¼æ•°
    B_s=1/box_N
    Bx=np.divide(Plot_black[0],B_s)
    Bx=np.ceil(Bx)
    By=np.divide(Plot_black[1],B_s)
    By=np.ceil(By)
    px=[]
    py=[]

# æ ¼å†…æ ‡è¯†
    
    for i in range(step_N):
        if BOX_k[int(Bx[i]),int(By[i])]==False:
            BOX_k[int(Bx[i]),int(By[i])] = True
            px.append((Bx[i]-0.5)/box_N)
            py.append((By[i]-0.5)/box_N)
  
    Zplt(px,py,Type="S",Ms=5000/box_N,Mm="s",Mc="r",Al=0.4,Size=True,La="åˆ†æ ¼")
    
    
widgets.interact(
    Determine_W_show,
    box_N=widgets.IntSlider(min=100, max=5000, step=100, value=500,description='N(è¾¹ç­‰åˆ†):'),
    step_N=widgets.IntSlider(min=10000, max=100000, step=1000, value=30000,description='M(è¡Œèµ°æ¬¡æ•°):'),
    Hurst=widgets.FloatSlider(min=0, max=1, step=0.0001, readout_format='.4f',value=0.5,description='H(èµ«æ–¯ç‰¹):'),
)
```

>### a. R/Såˆ†æâ€”â€”é‡æ ‡åº¦æ¶¨è½æå·®

<font size=3.5>**1**. è¾“å…¥æ•°æ®ä¸ºé•¿åº¦ä¸º $M$ çš„æ—¶é—´åºåˆ—æ•°æ®ã€‚å°†å…¶å–å¯¹æ•°å¹¶åšå·®åˆ†ï¼Œå˜æˆé•¿åº¦ä¸º $N=M-1$ çš„å¯¹æ•°å·®åˆ†åºåˆ—ï¼ˆæ³¨æ„è¯¥æ—¶é—´åºåˆ—æ•°æ®å‡ä¸ºæ­£æ•°ï¼‰ï¼š

<font size=3.5>$$N_i=\log(M_{i+1}/M_i),i=1,2,\cdots,M-1$$

<font size=3.5>åº”æ³¨æ„ï¼Œ $R/S$ ä¸èƒ½ç›´æ¥åº”ç”¨äºåˆ†æ•°å¸ƒæœ—è¿åŠ¨ä¹‹ä¸­ï¼Œåº”å–å…¶ä¸€é˜¶å·®åˆ†
    
<font size=3.5>**2**. å°†é•¿åº¦ä¸º $N$ çš„å¯¹æ•°æ”¶ç›Šç‡åºåˆ—ç­‰åˆ†ä¸º $A$ ä¸ªå­é›†ï¼Œæ¯ä¸ªå­é›†çš„é•¿åº¦ä¸º $N/A$ ã€‚è®¡ç®—æ¯ä¸ªå­é›†çš„å‡å€¼ï¼Œè®°ä¸º $e_a,a=1,2,\cdots,A$ã€‚åœ¨æ¯ä¸ªå­é›† $a$ å†…ï¼Œé€ä¸€è®¡ç®—å‰ $k$ ä¸ªç‚¹($k=1,2,\cdots,n$)ç›¸å¯¹è¯¥å­é›†å‡å€¼ $e_a$ çš„ç´¯ç§¯ç¦»å·®ï¼š
$$X_{k,a}=\sum_{i=1}^k(N_{i,a}-e_a),k=1,2,\cdots,n$$

<font size=3.5>**3**. è®¡ç®—æ¯ä¸ªå­é›† $a$ å†…å¯¹æ•°å¢é•¿ç‡åºåˆ—çš„æ³¢åŠ¨èŒƒå›´ $R_a$ ï¼Œå®ƒç­‰äºç´¯ç§¯ç¦»å·®æœ€å¤§å€¼å’Œæœ€å°å€¼çš„å·®å€¼ï¼š

<font size=3.5>$$R_a=\text{max}(X_{k,a})-\text{min}(X_{k,a}),1\leq k\leq n$$

<font size=3.5>**4**. è®¡ç®—æ¯ä¸ªå­é›† $a$ å†…å¯¹æ•°å¢é•¿ç‡åºåˆ—çš„æ ‡å‡†å·® $S_a$ ï¼Œå¯¹æ¯ä¸ªå­é›† $a$ å†…ï¼Œä½¿ç”¨å…¶æ ‡å‡†å·® $S_a$ å¯¹å…¶æ³¢åŠ¨èŒƒå›´ $R_a$ è¿›è¡Œæ ‡å‡†åŒ–ï¼Œå¾—åˆ°é‡æ ‡æå·® $R_a/S_a$ ã€‚

<font size=3.5>**5**. ä»ç¬¬äºŒæ­¥å¼€å§‹ï¼Œå¯¹äºé€‰å–çš„é•¿åº¦ $n$ ï¼Œæˆ‘ä»¬ä¸€å…±æœ‰ $A$ ä¸ªå­é›†ï¼Œå› æ­¤æœ‰ $A$ ä¸ªé‡æ ‡æå·®ã€‚å–å®ƒä»¬çš„å‡å€¼ä½œä¸ºè¯¥åŸå§‹å¯¹æ•°ä»·æ ¼åºåˆ—åœ¨é•¿åº¦ä¸º $n$ çš„æ—¶é—´è·¨åº¦ä¸Šçš„é‡æ ‡æå·®ï¼Œè®°ä¸º $(R/S)_n$ï¼š
$$\begin{aligned}(R/S)_n&=\frac{1}{A}\sum_{a=1}^A\frac{R_a}{S_a}\end{aligned}$$

<font size=3.5>**6**. å¢å¤§ $n$ çš„å–å€¼ï¼Œå¹¶é‡å¤å‰å…­æ­¥ï¼Œå¾—åˆ°ä¸åŒé•¿åº¦ $n$ çš„æ—¶é—´è·¨åº¦ä¸Šå¯¹æ•°åºåˆ—çš„é‡æ ‡æå·® $(R/S)_n$ ã€‚

<font size=3.5>**7**. Hurst æŒ‡æ•° $H$ æ˜¯æè¿° $(R/S)_n$ å’Œ $n^H$ çš„æ­£æ¯”å…³ç³»ï¼Œå³ï¼š

<font size=3.5>$$(R/S)_n=C\times n^H$$

<font size=3.5>**8**. å¯¹ $(R/S)_n$ å’Œ $n^H$ è¿›è¡ŒåŒå¯¹æ•°å›å½’ï¼Œå³ä½¿ç”¨ $\text{log}(n)$ å¯¹ $\text{log}((R/S)_n)$
 è¿›è¡Œçº¿æ€§å›å½’ã€‚å›å½’æ–¹ç¨‹çš„æˆªè·å°±æ˜¯ä¸Šé¢å…³ç³»ä¸­çš„å¸¸æ•° $C$ ï¼Œè€Œæ–œç‡å°±æ˜¯ $Hurst$ æŒ‡æ•° 
 

>### b.ç»å…¸R/Såˆ†ææ³•

```python
def hurst_exponent_estimation_classics(X,real_h):
    """
    ç»å…¸æ–¹æ³•ä¼°ç®— Hurst æŒ‡æ•°
    
    å‚æ•°ï¼š
    x - æ—¶é—´åºåˆ—æ•°æ®
    step - æ­¥é•¿
    
    è¿”å›ï¼š
    ä¼°è®¡ Hurst æŒ‡æ•°è¿‡ç¨‹ä¸­ç»˜å›¾
    
    """    
    #ï¼ˆå¯¹æ•°ï¼‰å·®åˆ†

    X=np.diff(X)
    
    #å–å­åºåˆ—
    R_S = []
    n=int(len(X))
    #å­åºåˆ—é•¿åº¦
    N_step = np.arange(10,int(n/2))
#     number=np.array([2**i for i in range(2,10)])
#     N_step = np.array([int(i) for i in number])
    for step in N_step:
        N = len(X)
        N_del=0
        while (N-N_del)%step != 0:
            N_del+=1
        #ä¸ç­‰åˆ†æ—¶æŠ›å¼ƒåé¢éƒ¨åˆ†æ•°æ®ã€‚
        x=X[:N-N_del]
        #å­åºåˆ—æ•°
        A=int((N-N_del)/step)
        subset_A = np.split(x,A ,axis = 0)

        # æ¯ä¸ªå­åºåˆ—çš„å‡å€¼
        subset_mean = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„æ ‡å‡†å·®
        subset_S = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
        subset_R = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„ç´¯è®¡ç¦»å·®
        subset_deviation = np.zeros([A,N])
        subset_R_S = np.zeros(A)
        #æ±‚ç´¯è®¡ç¦»å·®å’Œç´¯è®¡ç¦»å·®çš„æå·®
        A_i=0
        for a in subset_A:
            subset_mean[A_i] = np.mean(a)
            subset_S[A_i] = np.std(a,ddof=1) 
            d_i = 0
            #å­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
            sum_deviation = 0
            for a_i in a:
                sum_deviation = sum_deviation + (a_i-subset_mean[A_i])
                subset_deviation[A_i,d_i]=sum_deviation
                d_i+=1
            deviation_max = np.max(subset_deviation[A_i])
            deviation_min = np.min(subset_deviation[A_i])
            subset_R[A_i] =deviation_max-deviation_min
            #è®¡ç®—é‡æ ‡æå·®
            subset_R_S [A_i] = subset_R[A_i]/subset_S[A_i]
            A_i+=1
        R_S.append(np.sum(subset_R_S)/A)
    
    H, B = np.polyfit(np.log(N_step),np.log(R_S), 1)
    
    X_show=np.linspace(np.log(np.min(N_step)),np.log(np.max(N_step)),100)
    
    plt.figure(figsize=(9,4))
    plt.xlabel("$log(n)$",size=12)
    plt.ylabel("log(R/S)",size=12)
    plt.title("ç»å…¸R/Sæ³•ä¼°ç®— Hurst æŒ‡æ•°",size=16)
    
    plt.scatter(np.log(N_step),np.log(R_S),s=1,label="å­åºåˆ—ä¸åŒåˆ†å‰²æ–¹å¼")
    plt.plot(X_show,X_show*H+B,"g",label=f"ä¼°è®¡ï¼šln(R/S)={int(H*1000)/1000} ln(n)+{int(B*1000)/1000}")
    plt.plot(X_show,X_show*real_h+B,'r',label=f"ç†è®ºï¼šln(R/S)={int(real_h*1000)/1000}ln(n)+{int(B*1000)/1000}")
    print("å®é™…ï¼š",real_h," ä¼°è®¡ï¼š",H," è¯¯å·®ï¼š"+str(int((-real_h+H)/real_h*10000)/100)+"%")
    plt.legend(loc="best")
    plt.show()
    return ()
```

>### c. R/S-ALä¿®æ­£

<font size=3.5>å¯¹äºè¾ƒå°çš„nï¼Œç™½å™ªå£°çš„R/Sç»Ÿè®¡åˆ†æä¸ç†è®ºä¸Šhurst=0.5çš„æ•°æ®è®¡ç®—æ‰€å¾—æ–œç‡æœ‰æ˜¾è‘—çš„åå·®ã€‚å› æ­¤é€šå¸¸å–è¿‘ä¼¼ï¼š

<font size=3.5>$$\mathbf{E}(R / S)_{n}=\left\{\begin{array}{ll}
\frac{n-\frac{1}{2}}{n} \frac{\Gamma\left(\frac{n-1}{2}\right)}{\sqrt{\pi} \Gamma\left(\frac{n}{2}\right)} \sum_{i=1}^{n-1} \sqrt{\frac{n-i}{i}} & \text { for } \quad n \leq 340 \\
\frac{n-\frac{1}{2}}{n} \frac{1}{\sqrt{n \frac{\pi}{2}}} \sum_{i=1}^{n-1} \sqrt{\frac{n-i}{i}} & \text { for } \quad n>340
\end{array}\right.$$

<font size=3.5>ç™½å™ªå£°çš„èµ«æ–¯ç‰¹æŒ‡æ•° H è®¡ç®—ä¸º0.5åŠ ä¸Š $(R/S)_n - E(R/S)_n$ çš„æ‹Ÿåˆæ–œç‡ã€‚ç»Ÿè®¡ç»“æœç”¨ R/S-A Lè¡¨ç¤ºã€‚

```python
def Er(n):
    if n<=340:
        return (n-1/2)/n*math.gamma((n-1)/2)/np.sqrt(np.pi)/math.gamma(n/2)*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )
    else:
        return (n-1/2)/n*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )/np.sqrt(n*np.pi/2)
```

```python
def hurst_exponent_estimation_AL(X,real_h):
    """
     ALçº æ­£æ–¹æ³•ä¼°ç®— Hurst æŒ‡æ•°
    
    å‚æ•°ï¼š
    x - æ—¶é—´åºåˆ—æ•°æ®
    step - æ­¥é•¿
    
    è¿”å›ï¼š
    ä¼°è®¡ Hurst æŒ‡æ•°è¿‡ç¨‹ä¸­ç»˜å›¾
    
    """    
    #ï¼ˆå¯¹æ•°ï¼‰å·®åˆ†

    X=np.diff(X)
    
    #å–å­åºåˆ—
    R_S = []
    R_S_AL=[]
    n=int(len(X))
    #å­åºåˆ—é•¿åº¦
    N_step = np.arange(10,int(n/2))
    for step in N_step:
        N = len(X)
        N_del=0
        while (N-N_del)%step != 0:
            N_del+=1
        #ä¸ç­‰åˆ†æ—¶æŠ›å¼ƒåé¢éƒ¨åˆ†æ•°æ®ã€‚
        x=X[:N-N_del]
        #å­åºåˆ—æ•°
        A=int((N-N_del)/step)
        subset_A = np.split(x,A ,axis = 0)

        # æ¯ä¸ªå­åºåˆ—çš„å‡å€¼
        subset_mean = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„æ ‡å‡†å·®
        subset_S = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
        subset_R = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„ç´¯è®¡ç¦»å·®
        subset_deviation = np.zeros([A,N])
        subset_R_S = np.zeros(A)
        #æ±‚ç´¯è®¡ç¦»å·®å’Œç´¯è®¡ç¦»å·®çš„æå·®
        A_i=0
        for a in subset_A:
            subset_mean[A_i] = np.mean(a)
            subset_S[A_i] = np.std(a,ddof=1) 
            d_i = 0
            #å­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
            sum_deviation = 0
            for a_i in a:
                sum_deviation = sum_deviation + (a_i-subset_mean[A_i])
                subset_deviation[A_i,d_i]=sum_deviation
                d_i+=1
            deviation_max = np.max(subset_deviation[A_i])
            deviation_min = np.min(subset_deviation[A_i])
            subset_R[A_i] =deviation_max-deviation_min
            #è®¡ç®—é‡æ ‡æå·®
            subset_R_S [A_i] = subset_R[A_i]/subset_S[A_i]
            A_i+=1
        R_S.append(np.sum(subset_R_S)/A)

        R_S_AL.append(Er(step))
    
    H, B = np.polyfit(np.log(N_step),np.log(R_S), 1)
#     print(N_step,R_S)
    X_show=np.linspace(np.log(np.min(N_step)),np.log(np.max(N_step)),100)
    
    plt.figure(figsize=(9,4))
    plt.xlabel("$log(n)$",size=12)
    plt.ylabel("log(R/S)",size=12)
    plt.title("R/S_ALæ³•ä¼°ç®— Hurst æŒ‡æ•°",size=16)
    
    plt.scatter(np.log(N_step),np.log(R_S),s=1,label="å­åºåˆ—ä¸åŒåˆ†å‰²æ–¹å¼")
    plt.plot(X_show,X_show*H+B,"g",label=f"ç»å…¸ä¼°è®¡ï¼šln(R/S)={int(H*1000)/1000} ln(n)+{int(B*1000)/1000}")
    if real_h !=0.5:
        print("æ³¨æ„æ­¤æ—¶HurstæŒ‡æ•°ä¸ä¸º0.5ã€‚") 
    H_AL,B_AL = np.polyfit(np.log(R_S_AL),np.log(R_S),1)
#     H_AL,B_AL = np.polyfit(R_S,R_S_AL,1)
#     H_AL=H_AL*0.5
#     H_AL = -np.mean(R_S_AL)+np.mean(R_S)
#     H_AL=np.mean(R_S)-np.mean(R_S_AL)
    H_AL=H_AL*0.5
        
    plt.plot(X_show,X_show*H_AL+B,"c",label=f"ALçº æ­£ï¼šln(R/S)={int(H_AL*1000)/1000} ln(n)+{int(B*1000)/1000}")
    plt.plot(X_show,X_show*real_h+B,'r',label=f"ç†è®ºï¼šln(R/S)={int(real_h*1000)/1000}ln(n)+{int(B*1000)/1000}")
    print("å®é™…ï¼š",real_h," ä¼°è®¡ï¼š",H_AL," è¯¯å·®ï¼š"+str(int((-real_h+H_AL)/real_h*10000)/100)+"%")
    plt.legend(loc="best")
    plt.show()
    return ()
```

>### d. R/S-Hnä¿®æ­£

<font size=3.5>æ ¹æ®ã€Dã€‘åœ¨R/S_ALåˆ†æçš„åŸºç¡€ä¸Šå¢åŠ æœ€åä¸€ä¸ªæ­¥éª¤ï¼Œå³è®¡ç®—ï¼š

<font size=3.5>$$\log H_n=\log\left(R/S\right)_n-\log E\left(R/S\right)_n+\log(n)/2$$

<font size=3.5>$$\log H_n=\log c+H\log n.$$

```python
def hurst_exponent_estimation_Hn(X,real_h):
    """
     ALçº æ­£æ–¹æ³•ä¼°ç®— Hurst æŒ‡æ•°
    
    å‚æ•°ï¼š
    x - æ—¶é—´åºåˆ—æ•°æ®
    step - æ­¥é•¿
    
    è¿”å›ï¼š
    ä¼°è®¡ Hurst æŒ‡æ•°è¿‡ç¨‹ä¸­ç»˜å›¾
    
    """    
    #ï¼ˆå¯¹æ•°ï¼‰å·®åˆ†

    X=np.diff(X)
    
    #å–å­åºåˆ—
    R_S = []
    logHn=[]
    n=int(len(X))
    #å­åºåˆ—é•¿åº¦
    N_step = np.arange(10,int(n/2))
    for step in N_step:
        N = len(X)
        N_del=0
        while (N-N_del)%step != 0:
            N_del+=1
        #ä¸ç­‰åˆ†æ—¶æŠ›å¼ƒåé¢éƒ¨åˆ†æ•°æ®ã€‚
        x=X[:N-N_del]
        #å­åºåˆ—æ•°
        A=int((N-N_del)/step)
        subset_A = np.split(x,A ,axis = 0)

        # æ¯ä¸ªå­åºåˆ—çš„å‡å€¼
        subset_mean = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„æ ‡å‡†å·®
        subset_S = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
        subset_R = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„ç´¯è®¡ç¦»å·®
        subset_deviation = np.zeros([A,N])
        subset_R_S = np.zeros(A)
        #æ±‚ç´¯è®¡ç¦»å·®å’Œç´¯è®¡ç¦»å·®çš„æå·®
        A_i=0
        for a in subset_A:
            subset_mean[A_i] = np.mean(a)
            subset_S[A_i] = np.std(a,ddof=1) 
            d_i = 0
            #å­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
            sum_deviation = 0
            for a_i in a:
                sum_deviation = sum_deviation + (a_i-subset_mean[A_i])
                subset_deviation[A_i,d_i]=sum_deviation
                d_i+=1
            deviation_max = np.max(subset_deviation[A_i])
            deviation_min = np.min(subset_deviation[A_i])
            subset_R[A_i] =deviation_max-deviation_min
            #è®¡ç®—é‡æ ‡æå·®
            subset_R_S [A_i] = subset_R[A_i]/subset_S[A_i]
            A_i+=1
        R_S.append(np.sum(subset_R_S)/A)
        logHn.append(np.log(np.sum(subset_R_S)/A)-np.log(Er(step))+np.log(step)/2)
    
    H, B = np.polyfit(np.log(N_step),np.log(R_S), 1)

    H_Hn,B_Hn = np.polyfit(np.log(N_step),logHn, 1)

    X_show=np.linspace(np.log(np.min(N_step)),np.log(np.max(N_step)),100)
    
    plt.figure(figsize=(9,4))
    plt.xlabel("$log(n)$",size=12)
    plt.ylabel("log(R/S)",size=12)
    plt.title("R/S_Hnæ³•ä¼°ç®— Hurst æŒ‡æ•°",size=16)
    
    plt.scatter(np.log(N_step),np.log(R_S),s=1,label="å­åºåˆ—ä¸åŒåˆ†å‰²æ–¹å¼")
    plt.scatter(np.log(N_step),logHn,s=1,label="Hnä¸‹å­åºåˆ—ä¸åŒåˆ†å‰²æ–¹å¼")
    plt.plot(X_show,X_show*H+B,"g",label=f"ç»å…¸ä¼°è®¡ï¼šln(R/S)={int(H*1000)/1000} ln(n)+{int(B*1000)/1000}")
    plt.plot(X_show,X_show*H_Hn+B_Hn,"c",label=f"Hnçº æ­£ï¼šln(R/S)={int(H_Hn*1000)/1000} ln(n)+{int(B_Hn*1000)/1000}")
    plt.plot(X_show,X_show*real_h+B,'r',label=f"ç†è®ºï¼šln(R/S)={int(real_h*1000)/1000}ln(n)+{int(B*1000)/1000}")
    print("å®é™…ï¼š",real_h," ä¼°è®¡ï¼š",H_Hn," è¯¯å·®ï¼š"+str(int((-real_h+H_Hn)/real_h*10000)/100)+"%")
    plt.legend(loc="best")
    plt.show()
    return ()
```

```python
def hurst_exponent_estimation(X,real_h):
    """
     3ç§æ–¹æ³•ä¼°ç®— Hurst æŒ‡æ•°
    
    å‚æ•°ï¼š
    x - æ—¶é—´åºåˆ—æ•°æ®
    step - æ­¥é•¿
    
    è¿”å›ï¼š
    ä¼°è®¡ Hurst æŒ‡æ•°è¿‡ç¨‹ä¸­ç»˜å›¾
    
    """    
    #ï¼ˆå¯¹æ•°ï¼‰å·®åˆ†

    X=np.diff(X)
    
    #å–å­åºåˆ—
    R_S = []
    R_S_AL=[]
    logHn=[]
    n=int(len(X))
    #å­åºåˆ—é•¿åº¦
    N_step = np.arange(3,int(n/2))
    for step in N_step:
        N = len(X)
        N_del=0
        while (N-N_del)%step != 0:
            N_del+=1
        #ä¸ç­‰åˆ†æ—¶æŠ›å¼ƒåé¢éƒ¨åˆ†æ•°æ®ã€‚
        x=X[:N-N_del]
        #å­åºåˆ—æ•°
        A=int((N-N_del)/step)
        subset_A = np.split(x,A ,axis = 0)

        # æ¯ä¸ªå­åºåˆ—çš„å‡å€¼
        subset_mean = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„æ ‡å‡†å·®
        subset_S = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
        subset_R = np.zeros(A)
        # æ¯ä¸ªå­åºåˆ—çš„ç´¯è®¡ç¦»å·®
        subset_deviation = np.zeros([A,N])
        subset_R_S = np.zeros(A)
        #æ±‚ç´¯è®¡ç¦»å·®å’Œç´¯è®¡ç¦»å·®çš„æå·®
        A_i=0
        for a in subset_A:
            subset_mean[A_i] = np.mean(a)
            subset_S[A_i] = np.std(a,ddof=1) 
            d_i = 0
            #å­åºåˆ—ç´¯è®¡ç¦»å·®çš„æå·®
            sum_deviation = 0
            for a_i in a:
                sum_deviation = sum_deviation + (a_i-subset_mean[A_i])
                subset_deviation[A_i,d_i]=sum_deviation
                d_i+=1
            deviation_max = np.max(subset_deviation[A_i])
            deviation_min = np.min(subset_deviation[A_i])
            subset_R[A_i] =deviation_max-deviation_min
            #è®¡ç®—é‡æ ‡æå·®
            subset_R_S [A_i] = subset_R[A_i]/subset_S[A_i]
            A_i+=1
        R_S.append(np.sum(subset_R_S)/A)
        R_S_AL.append(Er(step))
        logHn.append(np.log(np.sum(subset_R_S)/A)-np.log(Er(step))+np.log(step)/2)
    
    H, B = np.polyfit(np.log(N_step),np.log(R_S), 1)
    
#     H_AL,B_AL = np.polyfit(np.log(N_step),np.log(R_S)-np.log(R_S_AL),1)
#     H_AL,B_AL = np.polyfit(np.log(R_S),np.log(R_S_AL),1)
    H_AL,B_AL = np.polyfit(np.log(R_S_AL),np.log(R_S),1)
    H_AL=H_AL*0.5
    H_Hn,B_Hn = np.polyfit(np.log(N_step),logHn, 1)


    return np.array([H,H_AL,H_Hn])
```

```python
def Er(n):
    if n<=340:
        return (n-1/2)/n*math.gamma((n-1)/2)/np.sqrt(np.pi)/math.gamma(n/2)*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )
    else:
        return (n-1/2)/n*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )/np.sqrt(n*np.pi/2)
```

```python
def Determine_W_N(Hurst,step_N,box_S):# å¾—ä¿ç•™æ ¼æ•°
    """
    è®¡ç®—å½“å‰æ¡ä»¶ä¸‹è¦†ç›–è¯¥åˆ†å½¢éœ€è¦æœ€å°‘æ ¼å­æ•°
    
    å‚æ•°ï¼š
    step_N - æ—¶é—´åºåˆ—æ•°æ®é‡
    Hurst - Hurst æŒ‡æ•°
    box_S - ç›’å­è¾¹é•¿
    
    è¿”å›ï¼š
    sum - å¯¹åº”æ¡ä»¶ä¸‹è¦†ç›–è¯¥åˆ†å½¢éœ€è¦æœ€å°‘æ ¼å­æ•°
    
    """    
#     Source=np.diff(Determine_W(step_N,Hurst))
    Source=Determine_W(step_N,Hurst)
    
# ç¡®å®šæ ¼æ•°
    #ç›’å­è¾¹é•¿
    B_s = box_S
    Bx=np.divide(Source[0],B_s)
    Bx=np.ceil(Bx)
    By=np.divide(Source[1],B_s)
    By=np.ceil(By)
    Bx_min=int(np.min(Bx))
    By_min=int(np.min(By))
    box_Nx=np.max(Bx)-Bx_min
    box_Ny=np.max(By)-By_min
    box_N=int(max(box_Nx,box_Ny))
#     print(box_S,box_N,box_Nx,box_Ny)
    BOX_k=BOX(box_N+1)

# æ ¼å†…æ ‡è¯†
    num=0
    for i in range(step_N):
#         if BOX_k[int(Bx[i-1])-Bx_min,int(By[i-1])-By_min]==False:
#             BOX_k[int(Bx[i-1])-Bx_min,int(By[i-1])-By_min] = True
#             num=num+1
        if BOX_k[int(Bx[i])-Bx_min,int(By[i])-By_min]==False:
            BOX_k[int(Bx[i])-Bx_min,int(By[i])-By_min] = True
            num=num+1
    
    return num


def func(x,k,b):
    return(k*x+b)
```

```python
def weierstrass_function(x,Hurst_exponent,b):
    """
    è®¡ç®—éšæœº Weierstrass å‡½æ•°ï¼Œç”ŸæˆfBMæ•°æ®
    
    å‚æ•°ï¼š
    x - æ—¶é—´åºåˆ—
    Hurst_exponent - Hurst æŒ‡æ•°
    b - Weierstrass å‡½æ•°çš„ç±»å‘¨æœŸæ§åˆ¶å‚é‡
    
    è¿”å›ï¼š
    å¯¹åº”xçš„Weierstrass å‡½æ•°å€¼
    
    """
    s = 0
    random.seed()
    #å–å‰100é¡¹
    c=np.random.standard_normal(1000)
    d=np.random.uniform(0,2*np.pi,size=1000)
    for k in range(1000):

        s += c[k] * np.power(b,-k * Hurst_exponent) * np.sin(  np.power(b,k) * x + d[k])
    return s
```
